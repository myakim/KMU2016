```python

import requests
import lxml.html
from urllib.parse import urljoin
from konlpy.tag import Twitter, Kkma, Komoran, Hannanum
from collections import Counter # 집계를 위한 패키지 import


board_url = 'http://news.joins.com/brandnews?cloc=joongang|home|biz'
response = requests.get(board_url)
response.content.decode('utf8')

root = lxml.html.fromstring(response.content)
root.xpath('.//*[@id="content"]/div/ul/li')[0].text_content()

ktag = Kkma()
```

```python
for node in root.xpath('.//*[@id="content"]/div/ul/li/strong/a'):
    
    # 게시물 리스트 제목 출력
    print(node.text_content())

    # 게시물 본문 URL
    main_text_url = urljoin(board_url, node.attrib['href'])
    print(main_text_url)
    print()
    
    # 게시물 본문 조회 및 출력
    response_main = requests.get(main_text_url)
    root_main_text = lxml.html.fromstring(response_main.content)
    main_text = root_main_text.xpath('.//div[@id="article_body"]')[0].text_content()
    print(main_text)
    print('━'*30)
    wlist = []
    for word in ktag.pos(main_text):  # 형태소 분리하여 하나의 단어씩
        if word[1] == 'NNG':          # 명사인 경우에
            wlist.append(word[0])     # wlist에 추가
    wlist = Counter(wlist)            # 각 단어가 몇 개인지 count
    print(wlist.most_common())               # count 결과를 내림차순 정렬
    print('━'*30)
```

```python
# 단어별 건수 집계 (마지막 기사 하나만 사용)
from collections import Counter # 집계를 위한 패키지 import
wlist = []
for word in ktag.pos(main_text):  # 형태소 분리하여 하나의 단어씩
    if word[1] == 'NNG':          # 명사인 경우에
        wlist.append(word[0])     # wlist에 추가
wlist = Counter(wlist)            # 각 단어가 몇 개인지 count
wlist.most_common()               # count 결과를 내림차순 정렬
```
