# 데이터 불러오기

```python
import numpy
```

```python
# numpy.save('tdm_small.npy', tdm)
```

```python
tdm = numpy.load('tdm_small.npy')
```

```python
tdm = tdm.tolist()
```

```python
tdm
```

```python
with open('words_small.txt', encoding='utf8') as f:
    words = f.read().splitlines()
```

```python
words
```

# gensim 포맷으로 바꾸기

```python
from gensim.matutils import Sparse2Corpus
```

```python
corpus = Sparse2Corpus(tdm.T)  ## .T : Transpose = 행열 변환
```

```python
corpus
```

# Latent Dirichlet Allocation

[LdaModel 문서](https://radimrehurek.com/gensim/models/ldamodel.html)

```python
from gensim.models.ldamodel import LdaModel
```

```python
lda = LdaModel(corpus=corpus, num_topics=100, id2word=dict(enumerate(words)))
## num_topics : 주제가 **개 있다고 가정해라.
## id2word : 주제를 indexing 함. 숫자가 랜덤할 수 있음.
```

# Topic 보기

```python
[(word, '{:.3f}'.format(prob*100)) for word, prob in lda.show_topic(4,20)]
```

```python
lda.show_topic(0)
```

# 문서의 topic 확인

## 문서 변환

```python
doc = [(i, tdm[0, i]) for i in range(1000)]
```

## 문서에서 가장 많이 쓰인 단어들 보기

```python
from operator import itemgetter
```

```python
doc_words = [(words[i], n) for i, n in doc]
```

```python
sorted(doc_words, key=itemgetter(1), reverse=True)[:10]
```

## 문서의 topic

```python
lda.get_document_topics(doc)
```

```python
lda.show_topic(97)
```

```python
lda.show_topic(81)
```
