# 국민일보

```python
import requests
import lxml.html
from urllib.parse import urljoin

board_url = 'http://news.kmib.co.kr/article/list_all.asp?sid1=all&st=all'
response = requests.get(board_url)
#response.content.decode('utf8')

root = lxml.html.fromstring(response.content)
root.xpath('.//dl[@class="nws"]/dt/a')[0].text_content()
```

```python
for node in root.xpath('.//dl[@class="nws"]/dt/a'):
    
    # 게시물 리스트 제목 출력
    print(node.text_content())

    # 게시물 본문 URL
    main_text_url = urljoin(board_url, node.attrib['href'])
    print(main_text_url)
    print()
    
    # 게시물 본문 조회 및 출력
    response_main = requests.get(main_text_url)
    root_main_text = lxml.html.fromstring(response_main.content)
    main_text = root_main_text.xpath('//div[@id="articleBody"]')[0].text_content()
    print(main_text)
    print('━'*30)
```
